# 调用Tensorflow-serving服务

本案例主要演示如何用客户端调用我们所部署的模型。

```python
import os
os.chdir("../../")
```

```python
import os
import cv2
import time
import grpc
import numpy as np
import tensorflow as tf

from concurrent import futures
from tensorflow_serving.apis import predict_pb2
from tensorflow_serving.apis import prediction_service_pb2_grpc

from crnn.utils.accuracy import acc
from configs.config import params as param
from crnn.data.postprocess.decode import Decoder

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

font = FontProperties(fname='./tests/fonts/msyh.ttc', size=20)
```

## 获取图片及其对应标签

```python
def get_images_labels():
    images = [cv2.imread(os.path.join(param['test_dataset_path'], img)) for img in os.listdir(param['test_dataset_path']) if '.jpg' in img]
    labels_path = [os.path.join(param['test_dataset_path'], img) for img in os.listdir(param['test_dataset_path']) if '.txt' in img]
    labels = []
    for file in labels_path:
        with open(file,'r',encoding='utf8') as f:
            label = f.read()
        labels.append(label)
    return images, labels
```

```python
images, labels = get_images_labels()
```

## 接口调用方法

```python
def text_ocr_grpc(image_list, ip='192.168.46.230', port='7500'):
    '''
    矩形框文字识别
    '''
    #解码器
    decoder = Decoder(param)
    result_list = []
    #grpc配置
    options = [('grpc.max_send_message_length', 1000 * 1024 * 1024), 
               ('grpc.max_receive_message_length', 1000 * 1024 * 1024)]
    channel = grpc.insecure_channel('{0}:{1}'.format(ip, port), options = options)
    stub = prediction_service_pb2_grpc.PredictionServiceStub(channel)
    request = predict_pb2.PredictRequest()
    request.model_spec.name = 'chinese'
    request.model_spec.signature_name = 'serving_default'
    imgH, imgW, imgC=(32, 4000, 3)
    #循环遍历每个图片
    for img in image_list:
        image = img.astype('float32')
        resized_image = tf.image.resize(image, [imgH, imgW],preserve_aspect_ratio=True)
        resized_image = resized_image / 255
        resized_image -= 0.5
        resized_image /= 0.5
        padding_im = tf.image.pad_to_bounding_box(resized_image,0,0,imgH, resized_image.shape[1])
        padding_im = np.expand_dims(padding_im, axis=0)
        request.inputs['input_1'].CopyFrom(tf.make_tensor_proto(np.asarray(padding_im)))
        response = stub.Predict(request, 1.0)
        predictions = tf.make_ndarray(response.outputs["output_1"])
        y_pred = decoder.decode(predictions, method='greedy')
        result_list.append(y_pred[0])
    channel.close()
    return result_list
```

## 调用接口

```python
start = time.time()
result = text_ocr_grpc(images[:100])
print('识别时间：', time.time()-start)
```

```python
# 结果展示
for i,sentense in enumerate(result):
    #if len(labels[i])<100:
    print(i,'---------------------------------------')
    plt.figure(figsize=(32,100))
    plt.subplot(len(result),1,i+1)
    plt.imshow(images[i]/255)
    plt.xlabel('识别结果： '+sentense, fontproperties=font)
    plt.show()
        #print('真实标签： '+labels[i])
```
